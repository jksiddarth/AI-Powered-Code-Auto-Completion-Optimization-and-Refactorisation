````python
from transformers import pipeline
import torch

# Load an open-source transformer model for code generation
MODEL_NAME = "EleutherAI/gpt-neo-1.3B"  # Efficient model (~6GB VRAM required)
device = 0 if torch.cuda.is_available() else -1  # Use GPU if available, otherwise CPU

generator = pipeline("text-generation", model=MODEL_NAME, device=device)

def generate_code(prompt, max_length=500):
    """
    Generates complete Python code based on the given prompt using a transformer model.
    """
    formatted_prompt = f"### Instruction:\nProvide a complete, executable Python function with an example.\n### Task:\n{prompt}\n### Code:\n```python\n"

    response = generator(
        formatted_prompt,
        max_length=max_length,
        do_sample=True,
        temperature=0.5,  # Balanced randomness for creativity
        top_p=0.95,       # Nucleus sampling for better diversity
        eos_token_id=50256  # Ensures proper stopping at end of function
    )

    # Extract and clean the response
    generated_code = response[0]["generated_text"].split("```python\n")[-1].strip()

    return generated_code

def optimize_code(code_snippet):
    """
    Optimizes a given Python function by improving performance and efficiency.
    """
    prompt = f"### Instruction:\nOptimize the following Python code for efficiency and best practices.\n### Code:\n```python\n{code_snippet}\n```\n### Optimized Code:\n```python\n"

    response = generator(
        prompt,
        max_length=500,
        do_sample=True,
        temperature=0.5,
        top_p=0.95,
        eos_token_id=50256
    )

    optimized_code = response[0]["generated_text"].split("```python\n")[-1].strip()

    return optimized_code

def refactor_code(code_snippet):
    """
    Refactors a given Python function to improve readability, modularity, and maintainability.
    """
    prompt = f"### Instruction:\nRefactor the following Python code to improve readability and modularity.\n### Code:\n```python\n{code_snippet}\n```\n### Refactored Code:\n```python\n"

    response = generator(
        prompt,
        max_length=500,
        do_sample=True,
        temperature=0.5,
        top_p=0.95,
        eos_token_id=50256
    )

    refactored_code = response[0]["generated_text"].split("```python\n")[-1].strip()

    return refactored_code

def interactive_mode():
    """
    Provides an interactive AI-powered coding assistant.
    """
    print("\nüîπ AI Code Assistant: Generate, Optimize, and Refactor Code üîπ\n")

    while True:
        print("\nOptions:")
        print("1. Generate New Code")
        print("2. Optimize Existing Code")
        print("3. Refactor Existing Code")
        print("4. Exit")

        choice = input("Choose an option (1/2/3/4): ").strip()

        if choice == "1":
            prompt = input("\nEnter your code request (e.g., 'Write a function to sort a list using QuickSort'):\n")
            print("\nüîπ Generated Code:\n", generate_code(prompt))

        elif choice == "2":
            code_snippet = input("\nEnter the code you want to optimize:\n")
            print("\nüîπ Optimized Code:\n", optimize_code(code_snippet))

        elif choice == "3":
            code_snippet = input("\nEnter the code you want to refactor:\n")
            print("\nüîπ Refactored Code:\n", refactor_code(code_snippet))

        elif choice == "4":
            print("\nüöÄ Exiting AI Code Assistant. Happy coding!")
            break

        else:
            print("\n‚ùå Invalid choice. Please enter 1, 2, 3, or 4.")

if __name__ == "__main__":
    interactive_mode()
````

````python

# Gradio UI
with gr.Blocks(title="AI Code Assistant") as demo:
    gr.Markdown("# ü§ñ AI Code Assistant: Generate, Optimize, and Refactor Python Code")

    with gr.Tabs():
        with gr.TabItem("Generate Code"):
            gen_input = gr.Textbox(label="Enter your code prompt", placeholder="e.g., Write a function to sort a list using QuickSort")
            gen_output = gr.Code(label="Generated Python Code")
            gen_button = gr.Button("Generate Code")
            gen_button.click(fn=generate_code, inputs=gen_input, outputs=gen_output)

        with gr.TabItem("Optimize Code"):
            opt_input = gr.Textbox(lines=8, label="Paste your Python code to optimize")
            opt_output = gr.Code(label="Optimized Python Code")
            opt_button = gr.Button("Optimize Code")
            opt_button.click(fn=optimize_code, inputs=opt_input, outputs=opt_output)

        with gr.TabItem("Refactor Code"):
            ref_input = gr.Textbox(lines=8, label="Paste your Python code to refactor")
            ref_output = gr.Code(label="Refactored Python Code")
            ref_button = gr.Button("Refactor Code")
            ref_button.click(fn=refactor_code, inputs=ref_input, outputs=ref_output)

# Launch the app
demo.launch(share=True)
````
